```python
df=df.drop_duplicates()
df
```

```python
df['student_name']=df['student_name'].str.strip()
```

```python
df['topic']=df['topic'].str.strip()
```

```python
df['tool_used']=df['tool_used'].str.strip()
```

```python
df['assignment_score'] = df['assignment_score'].where(
    (df['assignment_score'] >= 0) & (df['assignment_score'] <= 100)
)
```

```python
df['attendance']=df['attendance'].str.strip().str.title()
```

```python
df['lab_completed']=df['lab_completed'].str.strip().str.title()
```

```python
df['api_used']=df['api_used'].str.strip().str.title()
```

```python

df['quiz_score'] = df['quiz_score'].where(
    (df['quiz_score'] >= 0) & (df['quiz_score'] <= 100)
)
```

```python
df['study_hours']=df['study_hours'].where(
    df['study_hours']>=0

)
```

```python
df['feedback_rating'] = df['feedback_rating'].where(
    (df['feedback_rating'] >= 1) & (df['feedback_rating'] <= 5)
)
```

```python
mean=df['assignment_score'].mean()
df['assignment_score'].fillna(mean,inplace=True)
mean=df['quiz_score'].mean()
df['quiz_score'].fillna(mean,inplace=True)
median=df['study_hours'].median()
df['study_hours'].fillna(median,inplace=True)
median=df['feedback_rating'].median()
df['feedback_rating'].fillna(median,inplace=True)
```

```python
conditions = [
    df['total_score'] >= 160,
    (df['total_score'] >= 120) & (df['total_score'] <= 159),
    (df['total_score'] >= 80) & (df['total_score'] <= 119),
    df['total_score'] < 80
]
choices = ['Excellent', 'Good', 'Average','Bad']
df['performance_level'] = np.select(conditions, choices, default=None)
```

```python
import seaborn as sns
sns.barplot(df,
            x='topic',
            y='total_score',
            estimator=np.mean
            )
plt.title('Average Total Score by Topic')
plt.xlabel('Topic')
plt.ylabel('Average Total Score')
plt.savefig('topic_score_chat.png')
plt.show()
```

```python
df['performance_level'].value_counts().plot(kind="bar",x='performance_level')
plt.title('Performance Level Distribution')
plt.xlabel('Performance Level')
plt.ylabel('Count')
plt.savefig('performance_level_chat.png')
plt.show()
```

```python
df['attendance'].value_counts().plot(kind='pie', figsize=(6, 6))
plt.title('Distribution of Attendance')
plt.ylabel('') # Hide the default 'count' label
plt.savefig('attendance_chat.png')
plt.show()
```

```python
sns.lineplot(df,
             x='topic',
             y='study_hours',
             estimator=np.mean)
plt.title('Average Study Hours by Topic')
plt.xlabel('Topic')
plt.ylabel('Average Study Hours')
plt.savefig('study_hours_chat.png')
plt.show()
```

```python
import json

# Create summary dictionary
summary = {
    "total_students": len(df),
    "average_assignment_score": df["assignment_score"].mean(),
    "average_quiz_score": df["quiz_score"].mean(),
    "average_study_hours":df["study_hours"].mean(),
    "most_common_topic": df["topic"].mode()[0]
}

# Save to JSON file
with open("summary.json", "w") as file:
    json.dump(summary, file, indent=4)

print("Summary saved to summary.json")
```

```python
import sqlite3
import pandas as pd
df=pd.read_csv("/content/cleaned_ai_learning_lab.csv")
df

# Connect to SQLite database
conn = sqlite3.connect('learning_lab.db')

# Save DataFrame to SQLite table
df.to_sql(
    name='student_learning',
    con=conn,
    if_exists='replace',  # replaces table if it already exists
    index=False
)

# Close connection
conn.close()

print("Data stored successfully in learning_lab.db")
```
